```python
#changing the string values M and B into numerical form for easy working
df["diagnosis"] = df["diagnosis"].map({'M':1, 'B':0})
df
```

```python
x = df.drop(["diagnosis"], axis=1)
x
```

```python
y = df["diagnosis"]
y
```

## Training the Dataset

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#dividing our data into test set and training set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#using StandardScaler to fit the training model and test set
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
x_train
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#applying RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
```

```python
#calculating the accuracy of our predicted results using accuracy_score
accuracy = accuracy_score(y_pred,y_test)
print(accuracy)
```

#### We created a Tumor Detection Model and used it to prediction of out test data with an accuracy of around 95%.
